## Conv Layer

### One Step Forward

```python
def step_forward(X, kernels, b, p=0,s=4):
    X_1 = X
    f = kernels.shape[0]
    n = X.shape[0]
    n_kernels = kernels.shape[-1]
    o_shape = int((n+2*p-f)/s+1)
    output = np.zeros((o_shape, o_shape, n_kernels))
    end_row, end_column = f, f
    if p>0:
        X_1 = np.zeros(np.pad(X[:,:,0], p).shape+(X.shape[2],))
        for i in range(X.shape[2]):
            X_1[:,:,i] = np.pad(X[:,:,i], p)
    
    for k in range(n_kernels):
        for i in range(0,n-f,s):
            for j in range(0, n-f ,s):
                output[i//s, j//s, k]=np.sum(X_1[i:end_row+i, j:end_column+j]*kernels[:,:,:,k])
        output[:,:,k] += b[k]
    z = np.copy(output)
    output[output<0]=0
    return output, z
```

### Pooling Layer

```python
def max_pooling(X, s=2, z=3):
    n=X.shape[0]
    o_output = int((n-z)/s+1)
    n_c = X.shape[2]
    output = np.zeros((o_output, o_output,n_c))
    for i in range(0,n-z,s):
        for j in range(0, n-z ,s):
            output[i//s, j//s] = np.max(X[i:z+i, j:z+j])
    z = np.copy(output)
    output[output<0]=0
    a = np.copy(output)
    return a, z
```

# Backward from the max pooling step

```python
def backward_from_max_pooling(X, deriv,s=2, z=3):
    n=X.shape[0] 
    n_c = X.shape[2]
    output = np.copy(X)
    for i in range(0,n-z,s):
        for j in range(0, n-z ,s):
            m = np.max(output[i:z+i, j:z+j])
            for x in range(2):
                for y in range(2):
                    if m != output[i+x, j+y]:
                        output[i+x, j+y] = 0
                    else:
                        output[i+x, j+y] = deriv[i//s, j//s]

    return output
```

### Fully Connected Layer

```python
def Flatten_Layer(X):
    output = X.flatten()
    return output
```

```python
def fully_connected_layer(X, w, b):
    X = np.dot(w,X)+b
    z = np.copy(X)
    X[X<0]=0
    a=np.copy(X)
    return a, z
```

### Softmax 

```python
def softmax(Z, w, b,C=4):
    z = w.dot(Z) + b
    t=np.exp(z)
    pred = t/np.sum(t)
    return pred, z
```
